```python
import pandas as pd

# Load the dataset
file_path = "Global_Superstore.csv"  # Replace with the actual path to your dataset
df = pd.read_csv(file_path)

# Split into Orders
orders_cols = ['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Product ID', 
               'Category', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit', 'Region']
orders_df = df[orders_cols]
orders_df.to_csv('Orders.csv', index=False)

# Split into Customers
customers_cols = ['Customer ID', 'Customer Name', 'Segment']
customers_df = df[customers_cols].drop_duplicates()
customers_df.to_csv('Customer.csv', index=False)

# Split into Region
region_cols = ['Region', 'State']
region_df = df[region_cols].drop_duplicates()
region_df.to_csv('Region.csv', index=False)

print("Files created: Orders.csv, Customer.csv, Region.csv")
```
